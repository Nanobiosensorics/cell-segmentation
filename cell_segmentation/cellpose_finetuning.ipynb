# Loading and visualizing the data

```python
import numpy as np
from glob import glob
from torch.utils.data import random_split, DataLoader, Dataset

class MicroDataset(Dataset):
  def __init__(self):
    self.files = glob("/content/micro_data/**/*_seg.npz", recursive=True)

  def __len__(self):
    return len(self.files)

  def __getitem__(self, idx):
    seg = np.load(self.files[idx])
    im_mic = seg["im_mic"]
    im_markers = seg["im_markers"]
    return im_mic, im_markers

dataset = MicroDataset()
train_size = int(0.98*len(dataset))
test_size = len(dataset) - train_size
train_dataset, test_dataset = random_split(dataset, [train_size, test_size])
```

```python
import matplotlib.pyplot as plt
import cv2

def get_contours(mic, markers):
  contours, _ = cv2.findContours(markers, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
  mic_with_contours = mic.copy()
  cv2.drawContours(mic_with_contours, contours, -1, (255, 0, 0), 2)
  return mic_with_contours

NUM_EXAMPLES = 4
ids = np.random.randint(0, len(dataset) + 1, size=NUM_EXAMPLES)

fig, ax = plt.subplots(nrows=NUM_EXAMPLES, ncols=2, figsize=(12, 12))
for i in range(NUM_EXAMPLES):
  mic, markers = dataset[ids[i]]
  mic_with_markers = get_contours(mic, markers)
  cv2.imwrite(f"/content/imgs/dataset{i}.png", mic)
  cv2.imwrite(f"/content/imgs/dataset{i}_markers.png", mic_with_markers)
  ax[i][0].set_title("Image")
  ax[i][0].imshow(mic)
  ax[i][0].axis('off')
  ax[i][1].set_title("Marker")
  ax[i][1].imshow(mic_with_markers)
  ax[i][1].axis('off')
```

# Training the cellpose model with our data

```python
import os
import tifffile

"""
Our dataset is too big for cellpose to train in one iteration because
it reads the whole dataset into memory.
For this purpose we are doing the training in iterations.
"""

train_iterations = 20
train_dir = "/content/train"
test_dir = "/content/test"

def create_cellpose_dataset(dataset, dir, iterations):
  if not os.path.exists(dir):
    os.mkdir(dir)
  boundaries = np.linspace(start=0, stop=(len(dataset)), num=iterations+1)
  for i in range(1, len(boundaries)):
    start, end = int(boundaries[i-1]), int(boundaries[i])
    iteration_dir = os.path.join(dir, str(i))
    if not os.path.exists(iteration_dir):
      os.mkdir(iteration_dir)
    save_cellpose_iteration(dataset, iteration_dir, start, end)

def save_cellpose_iteration(dataset, dir, start, end):
  for i in range(start, end):
    img, masks = dataset[i]
    tifffile.imwrite(os.path.join(dir, f"{i}.tif"), img)
    tifffile.imwrite(os.path.join(dir, f"{i}_masks.tif"), masks)

create_cellpose_dataset(train_dataset, train_dir, train_iterations)
create_cellpose_dataset(test_dataset, test_dir, 1)
```

```python
from cellpose import io, models, train, core

n_epochs =  100
initial_model = "cyto3"
model_name = "segmentation_model"
learning_rate = 0.1
weight_decay = 0.0001
channels = [0, 0]

logger = io.logger_setup()

use_GPU = core.use_gpu()
print(use_GPU)
model = models.CellposeModel(gpu=use_GPU, model_type=initial_model)

def train_iteration(iteration):
  output = io.load_train_test_data(
      os.path.join(train_dir, str(iteration)),
      os.path.join(test_dir, "1"))
  train_data, train_labels, _, test_data, test_labels, _ = output
  return train.train_seg(model.net, train_data=train_data,
                              train_labels=train_labels,
                              test_data=test_data,
                              test_labels=test_labels,
                              channels=channels,
                              save_path="/content/drive/MyDrive",
                              n_epochs=n_epochs,
                              learning_rate=learning_rate,
                              weight_decay=weight_decay,
                              SGD=True,
                              nimg_per_epoch=8,
                              model_name=model_name)

for i in range(1, train_iterations+1):
  train_iteration(i)
```

```python
from cellpose import models

model = models.CellposeModel(gpu=True,
                             pretrained_model="/content/drive/MyDrive/models/segmentation_model")
```

```python
import time

def dice_score(y_true, y_pred):
    intersection = np.logical_and(y_true, y_pred).sum()
    sum_y_true = y_true.sum()
    sum_y_pred = y_pred.sum()

    if sum_y_true + sum_y_pred == 0:
        return 1.0  # Both masks are empty

    return 2.0 * intersection / (sum_y_true + sum_y_pred)

avg_dice = 0.
avg_execution = 0.
detected_cells = 0
total_cells = 0

for im_mic, im_markers in test_dataset:
  start_time = time.time()
  masks, _, _ = model.eval(im_mic, diameter=None, channels=[0, 0])
  binary_predicted = (masks != 0).astype(np.uint8)
  binary_original = (im_markers != 0).astype(np.uint8)

  dice = dice_score(binary_original, binary_predicted)
  avg_dice += dice

  execution_time = time.time() - start_time
  avg_execution += execution_time
  print(np.max(masks), np.max(im_markers))
  detected_cells += np.max(masks)
  total_cells += np.max(im_markers)

avg_dice /= len(test_dataset)
avg_execution /= len(test_dataset)

print(f"The average dice score on the test dataset is: {avg_dice}")
print(f"The average execution time on the test dataset is: {avg_execution}")
print(f"The cell detection rate on the test dataset is: {detected_cells / total_cells}")
```
